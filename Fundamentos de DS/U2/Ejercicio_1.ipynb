```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
current_directory = os.getcwd()
file = os.path.join(current_directory, "datasets/flete-aereo-vacunas-covid19-al-2021-06-28.xlsx")
raw_data = pd.read_excel(file,skiprows=4)
raw_data
```

```python
step_1_data = raw_data.iloc[:-1]
step_1_data = step_1_data.dropna(how = 'all')
step_1_data
```

```python
conteo_vuelos = step_1_data['vuelo '].value_counts()

porcentajes = (conteo_vuelos / len(step_1_data)) * 100

plt.figure(figsize=(8, 8))
patches, texts, autotexts = plt.pie(porcentajes, labels=['']*len(porcentajes), autopct='%1.1f%%', startangle=90, labeldistance = 2)
plt.legend(patches, porcentajes.index, loc="best")
plt.title('Porcentaje de vuelos')
plt.show()
```

```python
tipo_dato = step_1_data['factura_moneda_monto'].dtype
tipo_dato
step_1_data = step_1_data.dropna()
```

```python
step_1_data['factura_moneda'] = ''
step_1_data['factura_monto'] = ''
step_1_data['factura_moneda'] = step_1_data['factura_moneda_monto'].str[:3]
step_1_data['factura_monto'] = step_1_data['factura_moneda_monto'].str[4:]
```

```python
step_1_data['factura_monto'] = step_1_data['factura_monto'].str.strip()
step_1_data['factura_monto'] = step_1_data['factura_monto'].str.replace(' ', '')
step_1_data['factura_monto'] = step_1_data['factura_monto'].str.replace('.', '')
step_1_data['factura_monto'] = step_1_data['factura_monto'].str.replace(',', '.')
step_1_data['factura_monto'] = step_1_data['factura_monto'].astype(float)
```

```python
data_eur = step_1_data[step_1_data['factura_moneda'] == 'EUR']
data_usd = step_1_data[step_1_data['factura_moneda'] == 'USD']

# Crear boxplots
plt.figure(figsize=(10, 6))
plt.boxplot([data_eur['factura_monto'], data_usd['factura_monto']], labels=['EUR', 'USD'])
plt.title('Boxplot de Monto de Facturas por Moneda')
plt.xlabel('Moneda')
plt.ylabel('Monto')
plt.show()
```
