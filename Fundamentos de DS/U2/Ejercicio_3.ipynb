```python
import pandas as pd
import numpy as np

def interpolacion_lineal(x, df):
    df_sorted = df.sort_values(by='x')
    x_values = df_sorted['x'].values
    y_values = df_sorted['y'].values
    y_interpolado = np.interp(x, x_values, y_values)
    return y_interpolado

```

```python
datos_df = pd.DataFrame({'x': [1, 2, 3, 10], 'y': [2, 3, 5, 6]})
x_valor = 4
y_interpolado = interpolacion_lineal(x_valor, datos_df)
print(f"El valor interpolado de y para x={x_valor} es {y_interpolado}.")
```
