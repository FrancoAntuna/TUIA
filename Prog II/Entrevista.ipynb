```python
import os
import re
from collections import Counter
```

```python
def tokenize(text):
    words = re.findall(r'\w+', text.lower()) #tokenizo por palabras, y usando lower para pasar a minusculas
    return words

def read_text(filename):
    with open(filename, 'r', encoding='utf-8') as file: #utf8 funciono correctamente
        text = file.read()
    return text
```

```python
file_exists = False
while not file_exists: #loopeo el while hasta encontrar un archivo valido
    filename = input("Ingrese el nombre del archivo: ")
    if filename:
        if os.path.exists(filename):
            print("El archivo existe.")
            file_exists = True
        else:
            print("El archivo no existe. Intente nuevamente.")

clean_text = read_text(filename)
text_tokenized = tokenize(clean_text)

ignorar = ["the", "and","in","of","is"] #lista de palabras que se ignoranen la busqueda

frecuencia_palabras = Counter(text_tokenized)
apariciones = {}
for palabra, frecuencia in frecuencia_palabras.items():
    if palabra in ignorar: #con esto salteo las iteraciones de las palabras que esten en ignorar
        continue
    apariciones[palabra] = frecuencia

apariciones_ordenadas = sorted(apariciones.items(), key=lambda x: x[1], reverse=True) #ordeno con sorted
for palabra, frecuencia in apariciones_ordenadas:
    print(f'Palabra: {palabra}, Frecuencia: {frecuencia}')

```
